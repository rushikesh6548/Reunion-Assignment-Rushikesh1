### Question 1 : 
Retrieve the top 5 customers who have made the highest average order amounts in the last 6 months. The average order amount should be calculated for each customer, and the result should be sorted in descending order.

```python
# Breaking down the query : 
# First we Join the tables to get all the information 

cur.execute("""
SELECT 
            C.cust_id, C.cust_name, C.cust_email,O.order_id,O.order_date as 'Date',
            OI.order_item_id,OI.product_id,OI.order_quantity,P.product_name,P.product_category,P.product_price
        
FROM        Customers C 

INNER JOIN  Orders O 

ON          C.cust_id = O.cust_id

INNER JOIN  Order_item OI

ON          OI.order_id= O.order_id

INNER JOIN  Products P 

ON          P.product_id = OI.product_id
""")

```

```python
# Converting the query to get a better view in the table format:
import pandas as pd

# Fetch all the results
results = cur.fetchall()

column_names = [desc[0] for desc in cur.description]

# Create a DataFrame using the results and column names
df = pd.DataFrame(results, columns=column_names)

df
```

2. Retrieve the list of customer whose order value is lower this year as compared to previous year

```python
# Let us save the Joined table as a View for futher use as well :
cur.execute(

"""

CREATE VIEW Joined_table AS

SELECT 
            C.cust_id, C.cust_name, C.cust_email,O.order_id,O.order_date as 'Date',
            OI.order_item_id,OI.product_id,OI.order_quantity,P.product_name,P.product_category,P.product_price
        
FROM        Customers C 

INNER JOIN  Orders O 

ON          C.cust_id = O.cust_id

INNER JOIN  Order_item OI

ON          OI.order_id= O.order_id

INNER JOIN  Products P 

ON          P.product_id = OI.product_id
    

"""



)
```

```python
cur.execute("""

SELECT * FROM Joined_table
""")

results = cur.fetchall()

column_names  = [desc[0] for desc in cur.description]
df = pd.DataFrame(results,columns= column_names)
df
```

```python
cur.execute("""

SELECT cust_name, YEAR(DATE) as 'YEAR', SUM(order_quantity*product_price)  as 'total_amountForYear'

FROM Joined_table 

GROUP BY cust_name,YEAR(DATE)

ORDER BY cust_name



""")

results = cur.fetchall()

column_names  = [desc[0] for desc in cur.description]
df = pd.DataFrame(results,columns= column_names)
df

```

```python
# Now furher showing only those customers whose order value is lower as compared to previous years using 2 cte's by seperating both years
cur.execute("""

WITH cte1 AS (
    SELECT
        cust_name,
        YEAR(DATE) AS 'YEAR',
        SUM(order_quantity * product_price) AS 'total_amountForYear'
    FROM
        Joined_table
    GROUP BY
        cust_name,
        YEAR(DATE)
    ORDER BY
        cust_name
),
cte2 AS (
    SELECT
        cust_name,
        YEAR,
        total_amountForYear,
        RANK() OVER (PARTITION BY cust_name ORDER BY total_amountForYear DESC) AS `Rank`
    FROM
        cte1

)

SELECT
    cust_name,
    YEAR,
    total_amountForYear,
    `Rank`
FROM
    cte2
WHERE
    `Rank` = 2 and YEAR = 2023;



""")

results = cur.fetchall()

column_names  = [desc[0] for desc in cur.description]
df = pd.DataFrame(results,columns= column_names)
df
```

### Question 4 
Retrieve the list of top 5 selling products. Further bifurcate the sales by product variants

```python
cur.execute("""

SELECT
    product_id,
    product_name,
    SUM(order_quantity) AS total_quantity
FROM
    Joined_table  
GROUP BY
    product_id,
    product_name
ORDER BY
    total_quantity DESC
LIMIT 5;




""")
results = cur.fetchall()

column_names  = [desc[0] for desc in cur.description]
df = pd.DataFrame(results,columns= column_names)
df
```

```python
# Sales by product variant : the highest category sales:
cur.execute("""

SELECT
    product_name,
    product_category,
    SUM(product_price*order_quantity) AS 'TotalSaleForCategory'
FROM
    Joined_table  
GROUP BY
    product_name,product_category
ORDER BY
    TotalSaleForCategory DESC;




""")
results = cur.fetchall()

column_names  = [desc[0] for desc in cur.description]
df = pd.DataFrame(results,columns= column_names)
df
```
