### We use a library called pandasql which allows us to use SQL queries on pandas dataframe

```python
import pandas as pd 
df = pd.read_csv('Time Series.csv')
```

```python
df
```

### First Solving for bot named 'Sharan'

```python
from pandasql import sqldf
```

```python
sqldf("""

WITH sorted_data AS (
    SELECT Name, Start, End, Activity,
           (SELECT COUNT(*) FROM df AS d WHERE d.Name = df.Name AND d.Start <= df.Start) AS row_num
    FROM df
    WHERE Name = 'Sharan'
),
work_periods AS (
    SELECT Name, Start, End, Activity,
           (SELECT COUNT(*) FROM sorted_data AS sd WHERE sd.Name = work_periods.Name AND sd.Start <= work_periods.Start) - 
           (SELECT COUNT(*) FROM sorted_data AS sd WHERE sd.Name = work_periods.Name) AS period_num
    FROM sorted_data AS work_periods
)
SELECT Name, MIN(Start) AS Start, MAX(End) AS End, GROUP_CONCAT(Activity, ', ') AS Activities
FROM work_periods
GROUP BY Name, period_num;


""")
```
