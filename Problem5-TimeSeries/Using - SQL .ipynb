### We use a library called pandasql which allows us to use SQL queries on pandas dataframe

```python
import pandas as pd 
df = pd.read_excel('Time Series.xlsx')
```

```python
df.head(10)
```

### First Solving for bot named 'Sharan'

```python
from pandasql import sqldf
```

```python
sqldf("""


 
WITH sorted_data AS (
    SELECT Name, Start, End, Activity
    FROM df
    WHERE Name = 'Sharan'
    ORDER BY Start
),
t2 AS (
    SELECT Name, Start, End, Activity, LEAD(Start) OVER() AS lead_start
    FROM sorted_data
),
t3 AS (
    SELECT Name, Start, End, lead_start, Activity,
        CASE
            WHEN End >= lead_start THEN 1
            ELSE 2
        END AS Label
    FROM t2
),
t4 AS (
    SELECT *,
        CASE
            WHEN Label = LAG(Label) OVER (ORDER BY Start) THEN 0
            ELSE 1
        END AS group_change
    FROM t3
),
t5 as (SELECT Name, Start, End, Activity, Label,
    SUM(group_change) OVER (ORDER BY Start) AS group_column
FROM t4)


SELECT Name, MIN(Start) AS Start, MAX(End) AS End, GROUP_CONCAT(Activity, ', ') AS Activities
FROM t5
GROUP BY Name, group_column

LIMIT 50
""")




```
